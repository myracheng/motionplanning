```python
import scipy
from scipy.signal import savgol_filter

# Make an x-variable for plotting
x = range(0, 2500)
# from pandas import rolling_median

window_size = 11

y_1 = scipy.signal.savgol_filter(rewards, window_size, 3)
y_2 = scipy.signal.savgol_filter(steps, window_size, 3)

y_3 = scipy.signal.savgol_filter(rewards2, window_size, 2)
y_4 = scipy.signal.savgol_filter(steps2, window_size, 2)

y_5 = scipy.signal.savgol_filter(rewards3, window_size, 2)
y_6 = scipy.signal.savgol_filter(steps3, window_size, 2)


y_7 = scipy.signal.savgol_filter(rewards4, window_size, 2)
y_8 = scipy.signal.savgol_filter(steps4, window_size, 2)


y_9 = scipy.signal.savgol_filter(rewards5, window_size, 2)
y_10 = scipy.signal.savgol_filter(steps5, window_size, 2)

y_11 = scipy.signal.savgol_filter(rewards6, window_size, 2)
y_12 = scipy.signal.savgol_filter(steps6, window_size, 2)
legends = ['actor rate 0.0001 critic rate 0.001', 'actor rate 0.00001 critic rate 0.001', 'actor rate 0.0001 critic rate 0.0001', 'RRT gamma = 0.9', 'actual rrt', 'every 50']
# Set up plot
p = bokeh.plotting.figure(x_axis_label='Episode',
                          y_axis_label='Reward')

# Populate the plot with glyphs

p.line(x, y_3, legend=legends[1], alpha=0.6)
# p.line(x, y_5, color='red', legend=legends[2], alpha=0.6)
p.line(x, y_1, line_width=1, color='orange', legend=legends[0], alpha=0.6)
# p.line(x, y_7, line_width=1, color='green', legend=legends[3], alpha=0.6)
p.line(x, y_9, line_width=1, color='green', legend=legends[4], alpha=0.6)
p.line(x, y_11, line_width=1, color='purple', legend=legends[5], alpha=0.6)


p2 = bokeh.plotting.figure(
                          x_axis_label='Episode',
                          y_axis_label='Steps')

p2.line(x, y_4,legend=legends[1], alpha=0.6)
# p2.line(x, y_6, color='red', legend=legends[2], alpha=0.6)
p2.line(x, y_2, line_width=1, color='orange',legend=legends[0], alpha=0.6)
# p2.line(x, y_8, line_width=1, color='green', legend=legends[3], alpha=0.6)
p2.line(x, y_10, line_width=1, color='green', legend=legends[4], alpha=0.6)
p2.line(x, y_12, line_width=1, color='purple', legend=legends[5], alpha=0.6)



p.legend.location = "bottom_right"

bokeh.io.show(p)
bokeh.io.show(p2)
```
