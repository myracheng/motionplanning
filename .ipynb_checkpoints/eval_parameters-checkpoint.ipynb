```python
import json
import ast

# legend = ['actor0.0001_critic0.001', 'actor0.00001_critic0.001', 'rrt_nodes_only', 'rrt_50']
legend = {}
rewards = {}
steps = {}

text_file = open("results/0.0001_0.001_20190416-233743.json", "r")
lines = text_file.readlines()
splitfiles = lines[0].split('][')
splitfiles[0] = splitfiles[0] + ']'
splitfiles[1] = '[' + splitfiles[1]
rewards['actor0.0001_critic0.001'] = ast.literal_eval(splitfiles[0])
steps['actor0.0001_critic0.001'] = ast.literal_eval(splitfiles[1])
legend['actor0.0001_critic0.001'] = 'actor0.0001, critic0.001'

text_file = open("results/0.00001_0.001_20190417-103328.json", "r")
lines = text_file.readlines()
splitfiles = lines[0].split('][')
splitfiles[0] = splitfiles[0] + ']'
splitfiles[1] = '[' + splitfiles[1]
rewards['actor0.00001_critic0.001'] = ast.literal_eval(splitfiles[0])
steps['actor0.00001_critic0.001'] = ast.literal_eval(splitfiles[1])
legend['actor0.00001_critic0.001'] = 'actor0.00001, critic0.001'

# text_file = open("0.0001_0.001_20190507-224635.json", "r")
# lines = text_file.readlines()
# splitfiles = lines[0].split('][')
# splitfiles[0] = splitfiles[0] + ']'
# splitfiles[1] = '[' + splitfiles[1]
# rewards['rrt_nodes_only'] = ast.literal_eval(splitfiles[0])
# steps['rrt_nodes_only'] = ast.literal_eval(splitfiles[1])
# legend['rrt_nodes_only'] = 'rrt_nodes_only'

# text_file = open("50stepnodes0.0001_0.001_20190507-232045.json", "r")
# lines = text_file.readlines()
# splitfiles = lines[0].split('][')
# splitfiles[0] = splitfiles[0] + ']'
# splitfiles[1] = '[' + splitfiles[1]
# rewards['rrt_50'] = ast.literal_eval(splitfiles[0])
# steps['rrt_50'] = ast.literal_eval(splitfiles[1])
# legend['rrt_50'] = 'rrt_50'

text_file = open("obstacles_0.0001_0.001_20190510-135923.json", "r")
lines = text_file.readlines()
splitfiles = lines[0].split('][')
splitfiles[0] = splitfiles[0] + ']'
splitfiles[1] = '[' + splitfiles[1]
rewards['1_obstacle'] = ast.literal_eval(splitfiles[0])
steps['1_obstacle'] = ast.literal_eval(splitfiles[1])
legend['1_obstacle'] = '1_obstacle'
```

```python
# These are the necessary submodules
import bokeh.io
import numpy as np
import bokeh.plotting

# This is necesssary to display in the notebook
bokeh.io.output_notebook()
```

```python
import scipy
from scipy.signal import savgol_filter

# select a palette
from bokeh.palettes import Dark2_5 as palette
# itertools handles the cycling
import itertools  


# create a color iterator
colors = itertools.cycle(palette)    

# Make an x-variable for plotting
x = range(0, 1000)
# from pandas import rolling_median

window_size = 11


# Set up plot
p = bokeh.plotting.figure(x_axis_label='Episode',
                          y_axis_label='Reward')

p2 = bokeh.plotting.figure(
                          x_axis_label='Episode',
                          y_axis_label='Steps')

for name, color in zip(legend.keys(), colors):
    y_1 = scipy.signal.savgol_filter(rewards[name], window_size, 3)
    y_2 = scipy.signal.savgol_filter(steps[name], window_size, 3)

    p.line(x, y_1, line_width=1, color=color, legend=legend[name], alpha=0.6)
    p2.line(x, y_2, line_width=1, color=color,legend=legend[name], alpha=0.6)



p.legend.location = "bottom_right"

bokeh.io.show(p)
bokeh.io.show(p2)
```
