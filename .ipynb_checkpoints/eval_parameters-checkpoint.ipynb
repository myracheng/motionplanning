```python
# These are the necessary submodules
import bokeh.io
import numpy as np
import bokeh.plotting

# This is necesssary to display in the notebook
bokeh.io.output_notebook()
```

```python
import scipy
from scipy.signal import savgol_filter

# Make an x-variable for plotting
x = range(0, 2500)
# from pandas import rolling_median

window_size = 11

y_1 = scipy.signal.savgol_filter(rewards, window_size, 3)
y_2 = scipy.signal.savgol_filter(steps, window_size, 3)

# y_2 = np.log(y_2)

y_3 = rewards2
y_4 = steps2

y_5 = rewards3
y_6 = steps3

y_3 = scipy.signal.savgol_filter(y_3, window_size, 2)
y_4 = scipy.signal.savgol_filter(y_4, window_size, 2)
y_5 = scipy.signal.savgol_filter(y_5, window_size, 2)
y_6 = scipy.signal.savgol_filter(y_6, window_size, 2)


y_7 = scipy.signal.savgol_filter(rewards4, window_size, 2)
y_8 = scipy.signal.savgol_filter(steps4, window_size, 2)
y_7[y_7< -40000] = -40000
y_8[y_8>1000] = 1000

# y_4 = np.log(y_4)


# Set up plot
p = bokeh.plotting.figure(x_axis_label='Episode',
                          y_axis_label='Reward')
p2 = bokeh.plotting.figure(
                          x_axis_label='Episode',
                          y_axis_label='Steps')

# Populate the plot with glyphs

p.line(x, y_3, legend='actor rate 0.00001 critic rate 0.001', alpha=0.6)
p.line(x, y_5, color='red', legend='actor rate 0.0001 critic rate 0.0001', alpha=0.6)
p.line(x, y_1, line_width=1, color='orange', legend='actor rate 0.0001 critic rate 0.001', alpha=0.6)
p.line(x, y_7, line_width=1, color='green', legend='RRT', alpha=0.6)



p2.line(x, y_4, legend='actor rate 0.00001 critic rate 0.001', alpha=0.6)
p2.line(x, y_6, color='red', legend='actor rate 0.0001 critic rate 0.0001', alpha=0.6)
p2.line(x, y_2, line_width=1, color='orange', legend='actor rate 0.0001 critic rate 0.001', alpha=0.6)

p2.line(x, y_8, line_width=1, color='green', legend='RRT', alpha=0.6)


p.legend.location = "bottom_right"
p.legend.click_policy="mute"
# Show the plot
bokeh.io.show(p)
bokeh.io.show(p2)
```
